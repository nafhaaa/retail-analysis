# check for outliers

```python
#The outliers were checked for price per unit and total sales 
```

```python
data.rename(columns={'Price per Unit': 'price'}, inplace=True)
#renamed the column price per unit to price
```

```python
data.head()
```

```python
#remove outliers for price
q1=data["price"].quantile(0.25) #lower quartile
q3=data["price"].quantile(0.75) #upper quartile
ll=q1-1.5*(q3-q1)
ul=q3+1.5*(q3-q1)
data=data[(data["price"]>=ll )& (data["price"]<=ul)]
```

```python
data.shape #since the number of rows didnt change there were no outliers for price
```

```python
#remove outliers for Total Sales
q11=data["Total Sales"].quantile(0.25) #lower quartile
q31=data["Total Sales"].quantile(0.75) #upper quartile
ll1=q11-1.5*(q3-q1)
ul1=q31+1.5*(q3-q1)
data=data[(data["Total Sales"]>=ll1 )& (data["Total Sales"]<=ul1)]
```

```python
data.shape #its noted that there are outliers for the total sales and those were removed.
```

# check for misspellings

```python
#misspellings in the sales channel column
data['Sales Channel'].unique()
```

```python
def clean(data):
    if data["Sales Channel"]== "On line":
        return 'Online'
    if data["Sales Channel"]== "Online":
        return 'Online'
    if data["Sales Channel"]== "In-Store":
        return 'In-store'
    if data["Sales Channel"]== "In Store":
        return 'In-store'
```

```python
data["Sales Channel"]=data.apply(clean, axis=1)
```

```python
data
```

```python
pd.isnull(data).sum(axis=0)
```

```python
#under product category there is a misspelling for acessories and Footwear
def clean1(data):
    if data["Product Category"]== "Accessories":
        return 'Accessories'
    if data["Product Category"]== "Acessories":
        return 'Accessories'
    if data["Product Category"]== "Footwear":
        return 'Footwear'
    if data["Product Category"]== "Foot wear":
        return 'Footwear'
    if data["Product Category"]== "Clothing":
        return 'Clothing'
```

```python
data["Product Category"]=data.apply(clean1, axis=1)
data
```

```python
pd.isnull(data).sum(axis=0)
```

```python
def clean2(data):
    if data["day"]==1 or data["day"]==8:
        return 'Friday'
    elif data["day"]==11 or data["day"]==4:
        return 'Monday'
    elif data["day"]==2 or data["day"]==9:
        return 'Saturday'
    elif data["day"]==6:
        return 'Wednesday'
    elif data["day"]==5 or data["day"]==26:
        return 'Tuesday'
    elif data["day"]==7:
        return 'Thursday'
    else:
        return 'Sunday'
data["dayofweek"]=data.apply(clean2, axis=1) 
    
```

```python
def clean3(time):
    if 6 <= time.hour < 12:
        return 'Morning'
    if 12 <= time.hour < 18:
        return 'Evening'
    else:
        return 'Night'

data['time_interval'] = data['time'].apply(clean3)
```
